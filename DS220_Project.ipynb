Don't want to deal with NaNs

```python
anime_df.dropna(inplace = True)
anime_df.count()
```

# Don't want movies/specials/OVAs get rid of 1 episode animes

```python
anime_df['Episodes'] = pd.to_numeric(anime_df['Episodes'], errors='coerce')
anime_df = anime_df[anime_df['Episodes'] > 1]
```

```python
anime_df.head(20)
```

# What are the top 10 most popular anime series in the dataset?

```python
anime_df.sort_values(by='Popularity', ascending=True).head(10)
```

#Is there a relationship between rating and popularity? (LOWER POPULARITY IS A GOOD THING, MEANS ITS MORE POPULAR)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Score', y='Popularity', data=anime_df, alpha=0.7)
plt.title('Anime Popularity vs Rating')
```

# More popular anime tend to be higher rated, with a heavy presence around 7.5 and the popularity being in the top ~1000

```python
print(anime_df.columns)
```

#Anime became really popular in 2020. Did studios take advantage of that and start producing more animes since 2020? And if so, how well rated were they?

```python
anime_YEAR_df = anime_df.copy()
anime_YEAR_df["Release Date"] = anime_YEAR_df["Release Date"].str[-4:].astype(int)
```

```python
anime_YEAR_df
```

```python
plt.figure(figsize=(32, 8))
sns.boxplot(x='Release Date', y='Score', data=anime_YEAR_df)
```

#Ratings don't have a spike, meaning quality of anime didn't change much in 2020. But still, we're there more anime released in 2020 as compared to other years?

```python
anime_YEAR_df["Release Date"].value_counts()

yr_counts = anime_YEAR_df["Release Date"].value_counts().reset_index()
yr_counts.columns = ['Release Date', 'count']

plt.figure(figsize=(33, 10))
sns.barplot(x='Release Date', y='count', data=yr_counts)
```

#2018 had highest count of anime released, no relation for covid and anime released

#How does episode count correlate with an anime's score? Do viewers prefer shorter or longer series?

```python
plt.figure(figsize=(32, 8))
sns.scatterplot(x='Episodes', y='Score', data=anime_YEAR_df)
plt.locator_params(axis='x', nbins=20)
```

```python
anime_YEAR_df[(anime_YEAR_df["Release Date"] == 1999) & (anime_YEAR_df["Episodes"] >= 200)]
```

#Hm seems like One Piece is not in this data set. Weird

```python
anime_YEAR_df[['Episodes', 'Score']].corr()
```

```python
anime_YEAR_df['Episode_Bin'] = pd.cut(anime_YEAR_df['Episodes'], bins=[0, 12, 50, 100, 200, 500, 2000])
sns.boxplot(x='Episode_Bin', y='Score', data=anime_YEAR_df)
```

#**Very small** positive correlation for length of anime vs rating

# Has the average number of episodes per anime changed over the years? (As in anime that was released in a given year, how many episodes did it finish with? *NOT* how many episdoes each anime produced in the year)

```python
ep_per_year = anime_YEAR_df.groupby('Release Date')['Episodes'].mean()
ep_per_year
```

```python
plt.figure(figsize=(32, 8))
ep_per_year_df = ep_per_year.reset_index()
sns.barplot(x='Release Date', y='Episodes', data=ep_per_year_df)
plt.locator_params(axis='x', nbins=20)
```

#1994 and 1979 were years that had animes produce the most episodes
#Makes sense as we get closer to 2025 the bars get lower, anime recently released tend to produce an episode a week for ~6-12 weeks
